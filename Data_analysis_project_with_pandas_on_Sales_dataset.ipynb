```python
import pandas as pd
```

```python
#print the list
import os
os.listdir("/content/sales")
```

```python
dataset=pd.DataFrame()
for x in os.listdir("/content/sales/"):
  if "Sales" in x:
    d=pd.read_csv(f"/content/sales/{x}")
    dataset=pd.concat([dataset,d])
```
